```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = int(input("Pick a number "))
iterations = [x]
count = 0
x_axis = []
while(x != 1):
    if((x%2)==0):
        x = x//2
    else:
        x = (x*3) + 1
    iterations.append(x)
    
for x in iterations:
    count += 1
    x_axis.append(count)
    


print(iterations)
print(f' There were a total of {count} iteraitons')
# x_axis = count
y_axis = iterations
plt.plot(x_axis, y_axis);
```
